```python
import os
import ants
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
reg_path = "p1_subjects"

l1 = [x for x in os.listdir(reg_path) if x.endswith("_5.nii.gz")]
```

```python
l1
```

```python
out_path = os.path.join(reg_path, "slices")

for file in l1:
    file_path = os.path.join(reg_path, file)
    img = ants.image_read(file_path)
    
    plt.imshow(np.rot90(img.numpy()[:,:,128]), cmap="Greys_r")
    
    root = file.split(".")[0]
    out_file = os.path.join(out_path, f"{root}.png")
    plt.savefig(out_file)
```
