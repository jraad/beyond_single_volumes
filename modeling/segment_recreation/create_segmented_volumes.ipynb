### Import Libraries

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Define Paths

```python
research_dir = r"D:/school/research"
code_dir = os.path.join(research_dir, "code")
model_dir = os.path.join(code_dir, "explore_again", "models")
data_dir = os.path.join(research_dir, "data")
dhcp_rel2 = os.path.join(data_dir, "dhcp_rel2")
processed_dir = os.path.join(dhcp_rel2, "processed")

volume_dir = os.path.join(processed_dir, "volumes")
seg_dir = os.path.join(processed_dir, "segments")
seg_vol_dir = os.path.join(processed_dir, "volume_segments")

l1_dir = os.path.join(volume_dir, "l1")
l5_dir = os.path.join(volume_dir, "l5")

l1_seg_dir = os.path.join(seg_dir, "l1")
l5_seg_dir = os.path.join(seg_dir, "l5")

l1_seg_vol_dir = os.path.join(seg_vol_dir, "l1")
l5_seg_vol_dir = os.path.join(seg_vol_dir, "l5")
```

### Get Masked T2 Volumes and Save

```python
%%time

level = 1
session_list = [x for x in os.listdir(os.path.join(volume_dir, f"l{level}")) if x.endswith("T2w.npy")]

for session in session_list[158:]:
    img_path = os.path.join(volume_dir, f"l{level}", session)
    img = np.load(img_path)
    
    session_name = session.split("_T2w")[0]
    seg_path = os.path.join(seg_dir, f"l{level}", f"{session_name}_seg.npy")
    seg = np.load(seg_path)
    
    seg_split = []
    for i in range(1, 10):
        seg_single = seg.copy()
        seg_single[seg_single != i] = 0
        seg_single[seg_single != 0] = 1
        seg_mask = img * seg_single
        seg_split.append(seg_mask)
    seg_split = np.array(seg_split)
    
    seg_split_path = os.path.join(seg_vol_dir, f"l{level}", f"{session_name}_seg_T2w.npy")
    np.save(seg_split_path, seg_split)

    print(f"Finished processing {session}")
```

```python
fig, ax = plt.subplots(1, 9, figsize=(20,20))

for i in range(0, 9):
    ax[i].imshow(seg_split[i][:,:,128])
```
